```python
from prophet import Prophet

def forecast_country(country_name, periods=10):
    
    country_df = df[df["country"] == country_name]

    if len(country_df) < 17:
        return None
    
    prophet_df = country_df[["year", "co2_per_capita"]]
    prophet_df.columns = ["ds", "y"]
    
    prophet_df["ds"] = pd.to_datetime(prophet_df["ds"], format="%Y")
    
    model = Prophet()
    model.fit(prophet_df)
    
    future = model.make_future_dataframe(periods=periods, freq='Y')
    forecast = model.predict(future)
    
    return model, forecast
```

```python
model, forecast = forecast_country("India", 10)
```

```python
model.plot(forecast)
```

```python
import numpy as np

def risk_analysis(country_name):
    
    country_df = df[df["country"] == country_name]
    ts = country_df["co2_per_capita"].values
    
    recent_growth = ((ts[-1] - ts[-5]) / ts[-5]) * 100
    
    volatility = np.std(ts[-10:])
    
    if recent_growth > 10:
        risk = "High"
    elif recent_growth > 3:
        risk = "Medium"
    else:
        risk = "Low"
        
    return {
        "recent_growth_%": round(recent_growth,2),
        "volatility": round(volatility,2),
        "risk_level": risk
    }
```

```python
risk_analysis("India")
```

```python
def what_if_simulation(country_name, reduction_percent):
    
    model, forecast = forecast_country(country_name, 10)
    
    predicted = forecast["yhat"].values[-10:]
    
    adjusted = predicted * (1 - reduction_percent/100)
    
    return predicted, adjusted
```
